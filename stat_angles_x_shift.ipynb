```python
def remove_outliers_iqr(data):
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    return [v for v in data if lower <= v <= upper]

x_filtered = remove_outliers_iqr(x)
angles_filtered = remove_outliers_iqr(angles)
```

```python
# Help me plot x
plt.figure(figsize=(10, 5))
plt.plot(x_filtered, marker='o', linestyle='-', color='b')
plt.title('Plot of x values')
plt.xlabel('Index')
plt.ylabel('Value')
plt.grid(True)
plt.xticks(range(len(x_filtered)), rotation=45)
plt.tight_layout()
plt.show()
```
